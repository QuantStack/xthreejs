```python
def m2f(m, f):
    f[0, :, :] =  + 0.2500000000000000*m[0, :, :] + 0.1250000000000000*m[1, :, :] + 0.2500000000000000*m[3, :, :]
    f[1, :, :] =  + 0.2500000000000000*m[0, :, :] + 0.1250000000000000*m[2, :, :] - 0.2500000000000000*m[3, :, :]
    f[2, :, :] =  + 0.2500000000000000*m[0, :, :] - 0.1250000000000000*m[1, :, :] + 0.2500000000000000*m[3, :, :]
    f[3, :, :] =  + 0.2500000000000000*m[0, :, :] - 0.1250000000000000*m[2, :, :] - 0.2500000000000000*m[3, :, :]
    f[4, :, :] =  + 0.2500000000000000*m[4, :, :] + 0.1250000000000000*m[5, :, :] + 0.2500000000000000*m[7, :, :]
    f[5, :, :] =  + 0.2500000000000000*m[4, :, :] + 0.1250000000000000*m[6, :, :] - 0.2500000000000000*m[7, :, :]
    f[6, :, :] =  + 0.2500000000000000*m[4, :, :] - 0.1250000000000000*m[5, :, :] + 0.2500000000000000*m[7, :, :]
    f[7, :, :] =  + 0.2500000000000000*m[4, :, :] - 0.1250000000000000*m[6, :, :] - 0.2500000000000000*m[7, :, :]
    f[8, :, :] =  + 0.2500000000000000*m[8, :, :] + 0.1250000000000000*m[9, :, :] + 0.2500000000000000*m[11, :, :]
    f[9, :, :] =  + 0.2500000000000000*m[8, :, :] + 0.1250000000000000*m[10, :, :] - 0.2500000000000000*m[11, :, :]
    f[10, :, :] =  + 0.2500000000000000*m[8, :, :] - 0.1250000000000000*m[9, :, :] + 0.2500000000000000*m[11, :, :]
    f[11, :, :] =  + 0.2500000000000000*m[8, :, :] - 0.1250000000000000*m[10, :, :] - 0.2500000000000000*m[11, :, :]

def f2m(f, m):
    m[0, :, :] =  + f[0, :, :] + f[1, :, :] + f[2, :, :] + f[3, :, :]
    m[1, :, :] =  + 4.0000000000000000*f[0, :, :] - 4.0000000000000000*f[2, :, :]
    m[2, :, :] =  + 4.0000000000000000*f[1, :, :] - 4.0000000000000000*f[3, :, :]
    m[3, :, :] =  + f[0, :, :] - f[1, :, :] + f[2, :, :] - f[3, :, :]
    m[4, :, :] =  + f[4, :, :] + f[5, :, :] + f[6, :, :] + f[7, :, :]
    m[5, :, :] =  + 4.0000000000000000*f[4, :, :] - 4.0000000000000000*f[6, :, :]
    m[6, :, :] =  + 4.0000000000000000*f[5, :, :] - 4.0000000000000000*f[7, :, :]
    m[7, :, :] =  + f[4, :, :] - f[5, :, :] + f[6, :, :] - f[7, :, :]
    m[8, :, :] =  + f[8, :, :] + f[9, :, :] + f[10, :, :] + f[11, :, :]
    m[9, :, :] =  + 4.0000000000000000*f[8, :, :] - 4.0000000000000000*f[10, :, :]
    m[10, :, :] =  + 4.0000000000000000*f[9, :, :] - 4.0000000000000000*f[11, :, :]
    m[11, :, :] =  + f[8, :, :] - f[9, :, :] + f[10, :, :] - f[11, :, :]

def transport(f):
    f[0, 1:, :] = f[0, :-1, :]
    f[1, :, 1:] = f[1, :, :-1]
    f[2, :-1, :] = f[2, 1:, :]
    f[3, :, :-1] = f[3, :, 1:]
    f[4, 1:, :] = f[4, :-1, :]
    f[5, :, 1:] = f[5, :, :-1]
    f[6, :-1, :] = f[6, 1:, :]
    f[7, :, :-1] = f[7, :, 1:]
    f[8, 1:, :] = f[8, :-1, :]
    f[9, :, 1:] = f[9, :, :-1]
    f[10, :-1, :] = f[10, 1:, :]
    f[11, :, :-1] = f[11, :, 1:]

def equilibrium(m):
    m[1, :, :] = m[4, :, :]
    m[2, :, :] = m[8, :, :]
    m[3, :, :] = 0.
    m[5, :, :] = 0.5*m[0, :, :]**2 + m[4, :, :]**2/m[0, :, :]
    m[6, :, :] = m[4, :, :]*m[8, :, :]/m[0, :, :]
    m[7, :, :] = 0.
    m[9, :, :] = m[4, :, :]*m[8, :, :]/m[0, :, :]
    m[10, :, :] = 0.5*m[0, :, :]**2 + m[8, :, :]**2/m[0, :, :]
    m[11, :, :] = 0.

def relaxation(m):
    m[1, :, :] += 2.0000000000000000*(m[4, :, :] - m[1, :, :])
    m[2, :, :] += 2.0000000000000000*(m[8, :, :] - m[2, :, :])
    m[3, :, :] *= (1. - 1.5000000000000000)
    m[5, :, :] += 1.5000000000000000*(0.5*m[0, :, :]**2 + m[4, :, :]**2/m[0, :, :] - m[5, :, :])
    m[6, :, :] += 1.5000000000000000*(m[4, :, :]*m[8, :, :]/m[0, :, :] - m[6, :, :])
    m[7, :, :] *= (1. - 1.2000000000000000)
    m[9, :, :] += 1.5000000000000000*(m[4, :, :]*m[8, :, :]/m[0, :, :] - m[9, :, :])
    m[10, :, :] += 1.5000000000000000*(0.5*m[0, :, :]**2 + m[8, :, :]**2/m[0, :, :] - m[10, :, :])
    m[11, :, :] *= (1. - 1.2000000000000000)

def onetimestep(m, f, fnew):
    transport(f)
    f2m(f, m)
    relaxation(m)
    m2f(m, f)
    
def periodic_x(f):
    f[:, 0, :] = f[:, nx-2, :]
    f[:, nx-1, :] = f[:, 1, :]    
    
def periodic_y(f):
    f[:, :, 0] = f[:, :, ny-2]
    f[:, :, ny-1] = f[:, :, 1]    
    
```

```python
import numpy as np

# problem initialization
dx = 1./32
rhoo = 1.
deltarho = 1.
Taille = 2.
xmin, xmax, ymin, ymax = -0.5*Taille, 0.5*Taille, -0.5*Taille, 0.5*Taille

x = np.linspace(xmin-dx/2, xmax + dx/2, Taille/dx+2)
x = x[np.newaxis, :]
y = np.linspace(ymin-dx/2, ymax + dx/2, Taille/dx+2)
y = y[:, np.newaxis]

def initialization_rho(x, y):
    return rhoo * np.ones((x.size, y.size)) + deltarho * ((x-0.5*(xmin+xmax))**2+(y-0.5*(ymin+ymax))**2 < 0.25**2)

nx = x.size
ny = y.size
nv = 12

rho = initialization_rho(x, y)

f = np.zeros((nv, nx, ny))
f_new = np.zeros_like(f)
m = np.zeros((nv, nx, ny))

for i in range(4):
    f[i, :, :] = .25*rho
f[4, :, :] = .125*rho**2
f[6, :, :] = -.125*rho**2
f[9, :, :] = .125*rho**2
f[11, :, :] = -.125*rho**2
```

```python
for i in range(100):
    onetimestep(m, f, f_new)
    periodic_x(f)
    periodic_y(f)
    #f, f_new = f_new, f
```
